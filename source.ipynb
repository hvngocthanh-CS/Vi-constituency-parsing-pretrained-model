```python
from nltk import Tree
import torch
import stanza
```

```python
save_model = '/content/drive/MyDrive/BerkeleyParsers/phobert_stanza_demo'
```

```python
vi_tree = stanza.Pipeline(lang='vi', model_dir=save_model, processors='tokenize, pos, constituency', treebank=True, package='default_accurate', use_gpu=False, verbose=None)
```

#Input a document (vi):

```python
vi_doc = 'Tôi rất thích học Xử lý Ngôn ngữ tự nhiên . Và nó giúp ích rất nhiều cho công việc của tôi sau này .'
```

#Output:

```python
# Extract constituency parse tree from the output
doc = vi_tree(vi_doc)
for sent in doc.sentences:
    print(sent.constituency)
```

#Input a sentence (vi):

```python
vi_sentence = 'Tôi là sinh viên khoa Khoa học máy tính.'
```

#Output:

```python
# Tokenize and encode the input sentence
sent = vi_tree(vi_sentence)
print(sent.sentences[0].constituency)
```
